```python
import numpy as np
import matplotlib.pyplot as plt

fs = 8000
dur = 0.5
f = 440

N = int(fs * dur)
n = np.arange(N)
x = np.sin(2*np.pi*f*n/fs)

plt.figure(figsize=(10,4))
plt.plot(n[:50], x[:50])
plt.title("440 HZ Sine (First 100 samples)")
plt.xlabel("n"); plt.ylabel("Amplitude"); plt.grid(True);
```

## Variables
- fs = It tells you how many samples per second are taken from a continuous signal to create the discrete-time signal. Here, fs = 8000 means 8000 samples per second. Should always be twice or *2 of frequency if violated seeing wrong signals
- dur = duration
- f = frequency. This is the frequency of the sine wave in Hz (cycles per second). Here, f = 440 means the sine wave oscillates 440 times per second (this is the standard musical pitch A4).

- N - total length of signal (every point)
- n = arranges

This code generates a 440 Hz sine wave (like the A4 note in music), sampled at 8000 samples per second, lasting 0.5 seconds. It then plots the first 50 samples so you can “zoom in” and clearly see the oscillations of the sine wave.
