```python
import numpy as np # type: ignore
import matplotlib.pyplot as plt # type: ignore

# Parameters
fs = 8000        # Sampling frequency
f = 440          # Signal frequency
duration = 0.5   # seconds

#time vetor
t = np.arange(0, duration, 1/fs)

# sinusoidal signal
# formula is x[n] = sin(2pi*f*n/fs). Its the General Formula for a Continuous-Time Sine Wave
x = np.sin(2 * np.pi * f * t) 

plt.plot(t[:200], x[:200])
plt.title("Original 440HZ wave")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()

```

```python
# Noisy Noise

strength_noise = 0.5

#randommize noise
x_noisy = x + strength_noise * np.random.randn(len(t))

plt.plot(t[:200], x_noisy[:200], label ="Noisy Signal")
plt.plot(t[:200], x[:200], label = "Orginal")
plt.title("Noisy and Original")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend()
plt.show()
```

```python
#Moving average filter

def moving_average(signal, L):
    return np.convolve(signal, np.ones(L)/L, mode='same')

L = 5
y = moving_average(x_noisy, L)

plt.plot(t[:200], x_noisy[:200], label="Noisy Signal", alpha=0.7)
plt.plot(t[:200], y[:200], label=f"Smoothed (L={L})", color="g")
plt.title("Moving Average Filter")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.legend()
plt.show()
```

```python
#compare Noise Strength
for ns in [0.1, 0.5, 1.0]:
    noisy = x + ns * np.random.randn(len(t))
    plt.plot(t[:200], noisy[:200], label=f"Noise strength={ns}")

plt.plot(t[:200], x[:200], label="Original", linewidth=2, color="k")
plt.legend()
plt.title("Effect of Increasing Noise Strength")
plt.show()
```

```python
#compare Window LEnggth

noisy = x + 0.5 * np.random.randn(len(t))

for L in [5,20,50]:
    y = moving_average(noisy, L)
    plt.plot(t[:200], y[:200], label=f"L={L}")

plt.plot(t[:200], noisy[:200], label="Noisy", alpha=0.6, color="r")
plt.legend()
plt.title("Effect of Window Length on Smoothing")
plt.show()
```
